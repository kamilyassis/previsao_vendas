# Limpeza e Análise dos Dados

```python
df_filt = df_exp[df_exp['sales'] > 0]

df_filt
```

```python
lojas_mv = df_filt.groupby('store_nbr')['date'].count()

lojas_mv = lojas_mv.sort_values(ascending=False)

lojas_mv
```

```python
df = df_filt[df_filt['store_nbr'] == 44]

df = df.sort_values(by='date')

df
```

```python
print('tem', df['family'].nunique(), 'tipos de itens')
```

```python
print('Quantidade total de vendas:', df['sales'].sum())
```

```python
catg_mv = df.groupby('family')['sales'].sum()

catg_mv = catg_mv.sort_values(ascending=False)

catg_mv
```

```python
produtos_mv = [catg_mv.index[i] for i in range(0, 5)]

produtos_mv
```

## Agrupando dados

```python
colunas_remover = ['id', 'store_nbr', 'onpromotion']

df = df.drop(colunas_remover, axis=1)

df
```

```python
df['date'] = pd.to_datetime(df['date'])

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['dates'] =  df['month'].astype(str)+ '-' + df['year'].astype(str)

colunas_remover = ['date', 'year', 'month']

df = df.drop(colunas_remover, axis=1)

df
```

```python
df['dates'] = pd.to_datetime(df['dates'])

df
```

```python
df_mensal = df.groupby([pd.Grouper(key='dates', freq='MS'), 'family']).sum()

df_mensal
```

```python
df_mensal = df_mensal.reset_index()
df_mensal
```

```python
df_mensal.rename(columns={'sales': 'quantity', 'dates': 'date'}, inplace=True)
```

# Novo csv para analise

```python
df_mensal.to_csv('dataset\corporacion-favorit\store44_mes.csv', index=False)
```
